```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generate the samples we need.
np.random.seed(10)
n = 20
m = 1000
dist_1 = np.random.normal(0, 1, n*m).reshape(n, m)
dist_2 = np.random.uniform(0, 1, n*m).reshape(n, m)
dist_3 = np.random.chisquare(1, n*m).reshape(n, m)
dist_4 = np.random.exponential(1, n*m).reshape(n, m)
dist_1_S = (np.mean(dist_1, axis=0)-0)*np.sqrt(n)
dist_2_S = (np.mean(dist_2, axis=0)-0.5)*np.sqrt(12)*np.sqrt(n)
dist_3_S = (np.mean(dist_3, axis=0)-1)/np.sqrt(2)*np.sqrt(n)
dist_4_S = (np.mean(dist_4, axis=0)-1)*np.sqrt(n)
```

```python
# Make the plot with four subplots

x = np.linspace(norm.ppf(0.001),norm.ppf(0.999), 1000)
plt.subplot(221)
(n_1, bins_1, patches_1) = plt.hist(dist_1_S, 45, normed=1)
plt.plot(x, norm.pdf(x),'r-')
plt.axis((-5, 5, 0, 0.5))
plt.title('Standard Normarl')
plt.subplot(222)
(n_2, bins_2, patches_2) = plt.hist(dist_2_S, 45, normed=1)
plt.plot(x, norm.pdf(x),'r-')
plt.axis((-5, 5, 0, 0.5))
plt.title('Uniform(0, 1)')
plt.subplot(223)
(n_3, bins_3, patches_3) = plt.hist(dist_3_S, 45, normed=1)
plt.plot(x, norm.pdf(x),'r-')
plt.axis((-5, 5, 0, 0.5))
plt.title('Chisquare k =1')
plt.subplot(224)
(n_4, bins_4, patches_4) = plt.hist(dist_4_S, 45, normed=1)
plt.plot(x, norm.pdf(x),'r-')
plt.axis((-5, 5, 0, 0.5))
plt.title(r'Exponential $\lambda = 1$')
_ =plt.tight_layout()
plt.show()
```

```python
def generate_random_walk(p, n):
    rand = np.random.binomial(1, p, n)
    step = rand
    w = np.zeros(n+1)
    for i in range(len(rand)):
        if rand[i] == 0:
            step[i] = -1
        w[i+1] = w[i] + step[i]
    return w
```

**(b)**

```python
#Generate random walk
np.random.seed(70)
n = 200
p_40 = generate_random_walk(0.4, n)
p_45 = generate_random_walk(0.45, n)
p_50 = generate_random_walk(0.5, n)
p_55 = generate_random_walk(0.55, n)
p_60 = generate_random_walk(0.6, n)

#Plot the trajectories of those random walk
x = np.arange(n+1)
plt.title('Trajectories of diferent random walk')
plt.plot(x, p_40, 'r-', label = '$ p=0.4$')
plt.plot(x, p_45, 'b-', label = '$ p=0.45$')
plt.plot(x, p_50, 'g-', label = '$ p=0.5$')
plt.plot(x, p_55, 'k-', label = '$ p=0.55$')
plt.plot(x, p_60, 'y-', label = '$ p=0.60$')         
_ = plt.legend(loc='best')
plt.show()
```
